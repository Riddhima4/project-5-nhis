#### Detect Outliers Using IQR

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))

outlier_counts = outliers.sum()

print("Outliers Count Per Column:\n")
print(outlier_counts[outlier_counts > 0].sort_values(ascending=False))
```

#### Handle Outliers (Capping)

```python
# Select all numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    LB = Q1 - 1.5 * IQR
    UB = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < LB, LB, df[col])
    df[col] = np.where(df[col] > UB, UB, df[col])


#Outliers in all numerical columns were capped using the IQR method,
#ensuring that extreme values do not distort the analysis while preserving all rows in the dataset.
```

```python
#again checking for outliers
# Select numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))

outlier_counts = outliers.sum()

print(" Outliers count per column after capping:\n")
print(outlier_counts[outlier_counts > 0].sort_values(ascending=False))

#outliers has been resolved
```

#### Elbow Method to Find Best k

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

sse = []
K = range(1, 10)
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km = km.fit(normalized)
    sse.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, sse, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE (Inertia)')
plt.title('Elbow Method to Determine Optimal k')
plt.grid(True)
plt.show()



# Insight:
#Look for the "elbow" (where the curve sharply changes slope).

#This is the optimal number of clusters.

#Confirms whether 3 groups is valid, or you need more/less segmentation.
```

## User Satisfaction Analysis

#### Calculate Engagement and Experience Scores

```python
#Engagement Score = Euclidean distance from low-engagement cluster
#Experience Score = Euclidean distance from worst-experience cluster
```

#### Prepare Engagement and Experience Datasets

```python
# Merge engagement (from full_engagement) and experience (df_exp) using MSISDN/Number
merged = pd.merge(
    full_engagement, 
    df_exp[['MSISDN/Number', 'Avg TCP Retrans (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)', 'ExperienceCluster']],
    on='MSISDN/Number'
)
```

#### Normalize Engagement Features

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

engage_cols = ['xDR Session Count', 'Dur. (ms)', 'Total Traffic (Bytes)']
engage_data = merged[engage_cols].fillna(0)

scaler_engage = MinMaxScaler()
engage_scaled = scaler_engage.fit_transform(engage_data)

# Get low-engagement cluster center
low_engage_center = engage_scaled[merged['EngagementCluster'] == merged['EngagementCluster'].min()].mean(axis=0)

# Compute Euclidean distance from low-engagement cluster
merged['EngagementScore'] = np.linalg.norm(engage_scaled - low_engage_center, axis=1)
```

#### Normalize Experience Features



```python
exp_cols = ['Avg TCP Retrans (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']
exp_data = merged[exp_cols].fillna(0)

scaler_exp = MinMaxScaler()
exp_scaled = scaler_exp.fit_transform(exp_data)

# Get worst-experience cluster center
worst_exp_center = exp_scaled[merged['ExperienceCluster'] == merged['ExperienceCluster'].min()].mean(axis=0)

# Compute Euclidean distance from worst-experience cluster
merged['ExperienceScore'] = np.linalg.norm(exp_scaled - worst_exp_center, axis=1)
```

#### Compute Final Satisfaction Score

```python
# Final Satisfaction Score = average of engagement and experience scores
merged['SatisfactionScore'] = (merged['EngagementScore'] + merged['ExperienceScore']) / 2

# Preview top satisfied users
top_satisfied = merged.sort_values(by='SatisfactionScore', ascending=False).head(10)
top_satisfied[['MSISDN/Number', 'SatisfactionScore']]
```

```python
#Insights:
#This method quantifies user satisfaction based on both behavior and network quality.
#Useful for identifying top customers, improving loyalty, and targeting at-risk users.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Use original features (engagement + experience)
features = ['xDR Session Count', 'Dur. (ms)', 'Total Traffic (Bytes)',
            'Avg TCP Retrans (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']

X = merged[features].fillna(0)
y = merged['SatisfactionScore']
```
